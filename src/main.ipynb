# Benchmark Search Methods on Different Map Difficulties

This section benchmarks multiple search methods on maps of various difficulty levels:
- **Search Methods**: IDA*, Full-Space A*, Push only A*, Push-IDA*, and Simulated Annealing
- **Map Difficulties**: Easy, Medium, Hard, Large, and Super Hard

```python
# Benchmark multiple solvers on maps of various difficulties and plot metrics
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import os
import re
import glob
from search_methods.ida_star import IDAStarSolver
from search_methods.a_star import AStarSolver
from search_methods.push_a_star import PushAStarSolver
from search_methods.push_idastar import PushIDAStarSolver
from search_methods.simulated_annealing import SimulatedAnnealingSolver
from sokoban.moves import BOX_LEFT, BOX_RIGHT, BOX_UP, BOX_DOWN
from sokoban.map import Map as SokMap

# Instrumentation: count states explored by monkey-patching filter_possible_moves
counter = {}
orig_filter = SokMap.filter_possible_moves
def instrument_filter(self, *args, **kwargs):
    counter['states_explored'] = counter.get('states_explored', 0) + 1
    return orig_filter(self, *args, **kwargs)
SokMap.filter_possible_moves = instrument_filter

# list of solver constructors
def get_solvers():
    return [
        ('IDA*', IDAStarSolver),
        ('A*', AStarSolver),
        ('Push-A*', PushAStarSolver),
        ('Push-IDA*', PushIDAStarSolver),
        ('SimAnneal', SimulatedAnnealingSolver)
    ]

# Maximum time in seconds to allow a solver to run
# If it takes longer, we'll time out and move to the next solver
MAX_SOLVE_TIME = 60  # 1 minute timeout

# Helper function to get difficulty level from filename
def get_difficulty(filename):
    if 'easy' in filename:
        return 'Easy'
    elif 'medium' in filename:
        return 'Medium'
    elif 'hard' in filename and not 'super' in filename:
        return 'Hard'
    elif 'large' in filename:
        return 'Large'
    elif 'super_hard' in filename:
        return 'Super Hard'
    else:
        return 'Unknown'

# Get all yaml files in the tests directory
test_maps = sorted(glob.glob('tests/*.yaml'))

# Group maps by difficulty for better organization
maps_by_difficulty = {}
for path in test_maps:
    difficulty = get_difficulty(path)
    if difficulty not in maps_by_difficulty:
        maps_by_difficulty[difficulty] = []
    maps_by_difficulty[difficulty].append(path)

print(f"Found {len(test_maps)} maps across {len(maps_by_difficulty)} difficulty levels:")
for difficulty, maps in maps_by_difficulty.items():
    print(f"- {difficulty}: {len(maps)} maps")

# prepare data collection
records = []

# Run all solvers on all maps
for difficulty, maps in maps_by_difficulty.items():
    print(f"\nBenchmarking {difficulty} maps...")
    for path in maps:
        map_name = os.path.basename(path).replace('.yaml', '')
        print(f"  Map: {map_name}")
        m = Map.from_yaml(path)
        # Show initial map state
        m.plot_map()
        for name, SolverClass in get_solvers():
            print(f"    Solver: {name}... ", end="")
            solver = SolverClass(m)
            counter['states_explored'] = 0
            start = time.time()
            try:
                # Set a timeout for the solve method
                import signal
                
                def timeout_handler(signum, frame):
                    raise TimeoutError(f"Solver took longer than {MAX_SOLVE_TIME} seconds")
                
                # Set the timeout
                signal.signal(signal.SIGALRM, timeout_handler)
                signal.alarm(MAX_SOLVE_TIME)
                
                # Try to solve the map
                moves = solver.solve()
                
                # Cancel the timeout if solve finishes before timeout
                signal.alarm(0)
            except TimeoutError as e:
                print(f"Timed out after {MAX_SOLVE_TIME} seconds")
                moves = None
            except Exception as e:
                print(f"Error: {e}")
                moves = None
            
            elapsed = time.time() - start
            states = counter['states_explored']
            
            if moves is None:
                print("No solution found")
                records.append({
                    'map': map_name,
                    'difficulty': difficulty,
                    'solver': name,
                    'solved': False,
                    'time': elapsed,
                    'moves': None,
                    'pulls': None,
                    'states_explored': states
                })
            else:
                # Visualize final solved state
                solved_map = m.copy()
                for mv in moves:
                    solved_map.apply_move(mv)
                solved_map.plot_map()
                pulls = sum(1 for mv in moves if mv in {BOX_LEFT, BOX_RIGHT, BOX_UP, BOX_DOWN})
                print(f"Solved in {elapsed:.2f}s with {len(moves)} moves, {pulls} pulls, {states} states explored")
                records.append({
                    'map': map_name,
                    'difficulty': difficulty,
                    'solver': name,
                    'solved': True,
                    'time': elapsed,
                    'moves': len(moves),
                    'pulls': pulls,
                    'states_explored': states
                })

# build DataFrame
df = pd.DataFrame(records)

# Display all results in one table
print("\nAll Results:")
pd.set_option('display.max_rows', None)
pd.set_option('display.max_columns', None)
pd.set_option('display.width', None)
print(df)
```

```python
# Visualize benchmark results

# Only include solved maps in visualizations
solved_df = df[df['solved'] == True].copy()

if len(solved_df) == 0:
    print("No maps were solved successfully by any solver.")
else:
    # Calculate success rate by solver and difficulty
    success_rate = df.groupby(['difficulty', 'solver'])['solved'].mean().reset_index()
    success_rate = success_rate.pivot(index='difficulty', columns='solver', values='solved')
    
    # Plot success rate
    plt.figure(figsize=(12, 6))
    success_rate.plot(kind='bar')
    plt.title('Success Rate by Solver and Difficulty')
    plt.xlabel('Difficulty')
    plt.ylabel('Success Rate')
    plt.ylim(0, 1.1)  # 0% to 100% with a little margin
    plt.legend(title='Solver')
    plt.grid(axis='y')
    plt.show()
    
    # plot: average time per solver and difficulty
    plt.figure(figsize=(14, 6))
    sns.barplot(data=solved_df, x='difficulty', y='time', hue='solver')
    plt.title('Average Solve Time by Solver and Difficulty')
    plt.xlabel('Difficulty')
    plt.ylabel('Time (seconds)')
    plt.legend(title='Solver')
    plt.grid(axis='y')
    plt.show()
    
    # plot: states explored by solver and difficulty
    plt.figure(figsize=(14, 6))
    sns.barplot(data=solved_df, x='difficulty', y='states_explored', hue='solver')
    plt.title('States Explored by Solver and Difficulty')
    plt.xlabel('Difficulty')
    plt.ylabel('States Explored')
    plt.yscale('log')  # Use log scale since values can vary by orders of magnitude
    plt.legend(title='Solver')
    plt.grid(axis='y')
    plt.show()
    
    # plot: pushes distribution by solver and difficulty
    plt.figure(figsize=(14, 6))
    sns.boxplot(data=solved_df, x='difficulty', y='pulls', hue='solver')
    plt.title('Distribution of Pulls by Solver and Difficulty')
    plt.xlabel('Difficulty')
    plt.ylabel('Number of Pulls')
    plt.legend(title='Solver')
    plt.grid(axis='y')
    plt.show()
    
    # plot: moves distribution by solver and difficulty
    plt.figure(figsize=(14, 6))
    sns.boxplot(data=solved_df, x='difficulty', y='moves', hue='solver')
    plt.title('Distribution of Total Moves by Solver and Difficulty')
    plt.xlabel('Difficulty')
    plt.ylabel('Number of Moves')
    plt.legend(title='Solver')
    plt.grid(axis='y')
    plt.show()
```

```python
# Detailed comparison of solvers

# Only include solved maps in comparison
if len(solved_df) > 0:
    # Create a table comparing the performance of each solver across different metrics
    solver_comparison = solved_df.groupby('solver').agg({
        'solved': 'count',
        'time': 'mean',
        'moves': 'mean',
        'pulls': 'mean',
        'states_explored': 'mean'
    }).reset_index()
    
    # Calculate success rate for each solver
    total_maps = df.groupby('solver')['map'].count().reset_index()
    total_maps.columns = ['solver', 'total_maps']
    solver_comparison = solver_comparison.merge(total_maps, on='solver')
    solver_comparison['success_rate'] = solver_comparison['solved'] / solver_comparison['total_maps']
    
    # Reorder columns for better readability
    solver_comparison = solver_comparison[['solver', 'success_rate', 'solved', 'total_maps', 'time', 'moves', 'pulls', 'states_explored']]
    
    # Rename columns for better readability
    solver_comparison.columns = ['Solver', 'Success Rate', 'Maps Solved', 'Total Maps', 'Avg Time (s)', 'Avg Moves', 'Avg Pulls', 'Avg States Explored']
    
    # Format numeric columns
    solver_comparison['Success Rate'] = solver_comparison['Success Rate'].map(lambda x: f"{x:.2%}")
    solver_comparison['Avg Time (s)'] = solver_comparison['Avg Time (s)'].map(lambda x: f"{x:.2f}")
    solver_comparison['Avg Moves'] = solver_comparison['Avg Moves'].map(lambda x: f"{x:.1f}")
    solver_comparison['Avg Pulls'] = solver_comparison['Avg Pulls'].map(lambda x: f"{x:.1f}")
    solver_comparison['Avg States Explored'] = solver_comparison['Avg States Explored'].map(lambda x: f"{x:.1f}")
    
    print("Overall Solver Performance:")
    display(solver_comparison)
    
    # Compare performance by difficulty level
    print("\nPerformance by Difficulty Level:")
    
    for difficulty in maps_by_difficulty.keys():
        difficulty_df = solved_df[solved_df['difficulty'] == difficulty]
        if len(difficulty_df) > 0:
            print(f"\n{difficulty} Maps:")
            difficulty_comparison = difficulty_df.groupby('solver').agg({
                'solved': 'count',
                'time': 'mean',
                'moves': 'mean',
                'pulls': 'mean',
                'states_explored': 'mean'
            }).reset_index()
            
            # Calculate success rate for each solver on this difficulty level
            total_maps_diff = df[df['difficulty'] == difficulty].groupby('solver')['map'].count().reset_index()
            total_maps_diff.columns = ['solver', 'total_maps']
            difficulty_comparison = difficulty_comparison.merge(total_maps_diff, on='solver')
            difficulty_comparison['success_rate'] = difficulty_comparison['solved'] / difficulty_comparison['total_maps']
            
            # Reorder and rename columns for better readability
            difficulty_comparison = difficulty_comparison[['solver', 'success_rate', 'solved', 'total_maps', 'time', 'moves', 'pulls', 'states_explored']]
            difficulty_comparison.columns = ['Solver', 'Success Rate', 'Maps Solved', 'Total Maps', 'Avg Time (s)', 'Avg Moves', 'Avg Pulls', 'Avg States Explored']
            
            # Format numeric columns
            difficulty_comparison['Success Rate'] = difficulty_comparison['Success Rate'].map(lambda x: f"{x:.2%}")
            difficulty_comparison['Avg Time (s)'] = difficulty_comparison['Avg Time (s)'].map(lambda x: f"{x:.2f}")
            difficulty_comparison['Avg Moves'] = difficulty_comparison['Avg Moves'].map(lambda x: f"{x:.1f}")
            difficulty_comparison['Avg Pulls'] = difficulty_comparison['Avg Pulls'].map(lambda x: f"{x:.1f}")
            difficulty_comparison['Avg States Explored'] = difficulty_comparison['Avg States Explored'].map(lambda x: f"{x:.1f}")
            
            display(difficulty_comparison)
        else:
            print(f"\n{difficulty} Maps: No maps were solved successfully by any solver.")
else:
    print("No maps were solved successfully by any solver.")
```
